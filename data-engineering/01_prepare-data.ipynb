```python
import json
import pandas as pd
newdata = pd.read_json('/Users/kiyas/Downloads/gpt-labelled.json',  orient='records', lines=True)
df = pd.read_json("hf://datasets/textminr/mn-ds/train.jsonl", lines=True)
```

```python
newdata = newdata[(newdata['category_level_1'] != "") & (newdata['category_level_2'] != "")]
cat1 = df['category_level_1'].unique()
cat2 = df['category_level_2'].unique()
```

```python
newdata['category_level_1'] = newdata['category_level_1'].apply(lambda x: x.lower())
newdata['category_level_2'] = newdata['category_level_2'].apply(lambda x: x.lower())
```

```python
newdata['category_level_1'] = newdata['category_level_1'].replace({
    'economy, business, and finance': 'economy, business and finance',
    'arts, culture, entertainment, and media': 'arts, culture, entertainment and media',
    'disaster, accident, and emergency incident': 'disaster, accident and emergency incident',
    'crime, law, and justice': 'crime, law and justice',
    'conflict, war, and peace': 'conflict, war and peace',
    'culture': 'arts, culture, entertainment and media',
    'social': 'society'
})
```

```python
newdata['category_level_2'] = newdata['category_level_2'].replace({
  'travel': 'leisure', 
  'personal finance': 'market and exchange', 
  'student loan forgiveness': 'economy',
       'financial aid':'economy', 
  'health policy': 'healthcare policy', 'fashion': 'lifestyle', 
  'real estate': 'market and exchange',
       'reading': 'lifestyle', 
  'sustainable architecture': 'environmental politics', 'environmental injustice': 'environmental politics',
       'mental health': 'health treatment', 
  'social security': 'economy', 'financial technology':'business information',
       'architecture':'arts and entertainment', 'housing market':'market and exchange', 'entertainment':'arts and entertainment', 'ageism': 'social condition'

})
df = pd.concat([df,newdata], axis=0)
```
